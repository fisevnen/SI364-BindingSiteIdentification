```python
indexs = np.load('part2-index.npy', allow_pickle=True)
indexs_list = indexs.tolist()
label_dic = np.load('part2-label.npy', allow_pickle=True)
```

```python
total_dic = {}
for i in indexs_list:
    print(''.join(i.split('-')) + '.pdb')
    if ''.join(i.split('-')) + '.pdb' in bench_list:
        print('pass')
        pass
    else:
        multi_feature = np.hstack((np.array([sasa_dic.item()[i]]).T, pssm_dic.item()[i], SS_dic.item()[i], np.array(RT_dic.item()[i]), np.array([label_dic.item()[i]]).T))
        # print(multi_feature.shape)
        total_dic[i] = multi_feature
```

```python
add_zero_dic = {}
add_zero_label_dic = {}
for i in indexs_list:
    multi_feature = np.hstack((np.array([sasa_dic.item()[i]]).T, pssm_dic.item()[i], SS_dic.item()[i], np.array(RT_dic.item()[i])))
    label = label_dic.item()[i]
    if multi_feature.shape[0] < 600:
        count = 600-multi_feature.shape[0]
        multi_feature = np.pad(multi_feature, ((0,count),(0,0)), 'constant', constant_values=(0,0))
        label = np.pad(label, (0,count))
    print(multi_feature.shape)
    print(label.shape)
    add_zero_dic[i] = multi_feature
    add_zero_label_dic[i] = label
```

```python
add_600_list = []
add_600_label_list = []
longer_600_dic = {}
for i in add_zero_dic.keys():
    if add_zero_dic[i].shape[0] == 600:
        add_600_list.append(add_zero_dic[i])
        add_600_label_list.append(add_zero_label_dic[i])
    else:
        longer_600_dic[i] = add_zero_dic[i]
```
