```python
from tqdm import tqdm
import shutil

single_chiain_list = []
for i in tqdm(file_list):
    name = i.split('-')[0]
    for k in index_contents:
        if name in k:
            single_chiain_list.append(k)
            # try:
            #     shutil.copy('IFP/'+i.split('-')[0]+'-IFP.txt', 'new-IFP-2/%s-%s.IFP' %(k.split('_')[0], k.split(',')[2]))

            # except:
            #     print(i)
            #     pass
```

```python
bench_uID_list = []
for i in tqdm(bench_list):
    pdbID = i[:4]
    chainID = i[4:5]
    for j in single_chiain_list:
        if pdbID == j[:4] and chainID == j.split(',')[2]:
            bench_uID_list.append(j)
```

```python
ben_geneID = [i.split(',')[3] for i in bench_uID_list]
bench_PDB_list = []
for i in ben_geneID:
    for j in single_chiain_list:
        if i in j:
            bench_PDB_list.append(j)
len(bench_PDB_list)
```

```python
list_for_cp = []
list_redownload = []
import shutil
import glob

for i in tqdm(single_chiain_list):
    single_chiain_file = glob.glob('../PDB/*' + i[:4] + '.1.' + i.split(',')[2] + '*')
    if len(single_chiain_file) == 0:
        list_redownload.append(i[:4]+'-'+i.split(',')[2])
    else:
        for k in single_chiain_file:
            list_for_cp.append(k)
            # shutil.copy(k, 'final_general_DB_single_chain/') 
```

```python
part1_bench_list = []
for i in tqdm(bench_PDB_list):
    single_chiain_file = glob.glob('../PDB/*' + i[:4] + '.1.' + i.split(',')[2] + '*')
    if len(single_chiain_file) == 0:
        list_redownload.append(i[:4]+'-'+i.split(',')[2])
    else:
        for k in single_chiain_file:
            part1_bench_list.append(k)
            # shutil.copy(k, 'final_general_DB_single_chain/') 
```
